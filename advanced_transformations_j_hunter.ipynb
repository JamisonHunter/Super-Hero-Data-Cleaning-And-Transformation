# Combining Data Frames

```python
# adding expand=True
info[["Hero", "Publisher"]] = info['Hero|Publisher'].str.split('|',expand=True)
```

```python
info.head()
```

```python
info = info.drop(columns = ["Hero|Publisher"])
info.head()
```

```python
info.sort_values("Hero")
```

```python
power.sort_values("hero_names")
```

```python
merged = pd.merge(info, power, left_on = "Hero", right_on = "hero_names")
merged.head()
```

We know now that the data frames are properly merged. We can now remove the "hero_names" column.

```python
merged = merged.drop(columns = ["hero_names"])
merged.head()
```

```python
# adding expand=True
merged[["Height", "Weight"]] = merged['Measurements'].str.split(',',expand=True)
merged.head()
```

```python
# Make a list of all characters to replace
to_replace = ['Weight','kg', '}', ':', '', "''", "'"]
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    merged['Weight'] = merged['Weight'].str.replace(char,'',regex=False)

merged['Weight'] = merged['Weight'].str.strip()
merged['Weight'] = merged['Weight'].str.replace(" ", "")
merged['Weight'] = merged['Weight'].astype('float64') 
    
merged['Weight'].head()
```

```python
# Make a list of all characters to replace
to_replace = ['Height','cm', '{', ':', '', "''", "'"]
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    merged['Height'] = merged['Height'].str.replace(char,'',regex=False)

merged['Height'] = merged['Height'].str.strip()
merged['Height'] = merged['Height'].str.replace(" ", "")
merged['Height'] = merged['Height'].astype('float64') 
    
merged['Height'].head()
```

```python
merged = merged.drop(columns = ["Measurements"])
```

```python
# checking data frame to make sure height and weight appear as floats
merged.head()
```

```python
merged["power_split"] = merged['Powers'].str.split(",")
merged["power_split"].head()
```

```python
## exploding the column of lists
exploded = merged.explode('power_split')
exploded[["power_split"]].head()
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['power_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    merged[col] = merged['Powers'].str.contains(col)
merged.head()
```

```python
merged = merged.drop(columns = ["Powers"])
merged.head()
```

I will find the average weight of heroes with super speed versus heroes without super speed.

```python
df = merged["Super Speed"] == True
merged[df].head()
```

```python
df[merged["Super Speed"]].unique()
```

```python
# Average weight of heroes with super speed
merged[df]["Weight"].mean()
```

```python
# Average weight of heroes without super speed
df = merged["Super Speed"] == False
merged[df]["Weight"].mean()
```

Next, I will find the average height of super heroes based on publisher.

```python
merged["Publisher"].unique()
```

```python
# creating a for loop in order to list average heights by publisher
for i in merged["Publisher"].unique():
    x = merged["Publisher"] == i
    print(f"The average seper hero height for {i} publisher is:")
    print(merged[x]["Height"].mean())
```
